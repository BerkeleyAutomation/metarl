```python
r = rt.MultiExperimentResults(paths) 
keys = r.data.keys()
# print  r.data['Voting-Eps5'].values()
```

```python
# data = dict((k, r.data[k][1]['return']) for k in keys)
data = dict((k, r.data[k].values()[0]['return']) for k in keys)
```

```python
def movingavg(arr, period, filt):
    return [filt(arr[i:i + period]) for i in range(len(arr) - period + 1)]
    
```

```python

import matplotlib.pyplot as plt
% matplotlib notebook
```

```python

# plt.figure()
per = 1
avg_data = dict((k, movingavg(v, per, np.median)) for k, v in data.items())
steps = r.data[keys[0]].values()[0]['learning_steps'][per - 1:]

for k, v in avg_data.items():
    plt.plot(steps,v,label=k)

plt.show()
plt.title("Evaluations".format(per))
plt.legend(loc=1, bbox_to_anchor=(0.1, 1), prop={'size':10})
```
